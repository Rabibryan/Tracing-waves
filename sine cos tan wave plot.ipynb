# Tracing of curves

sine, cosine and tan waves

```python
import numpy as np
import matplotlib.pyplot as plt
```

Sine wave

```python
time =np.arange(0,20,0.1)
amplitude =np.sin(time)
plt.plot(time, amplitude)
plt.title('Sine wave')
plt.xlabel('Time')
plt.ylabel('Amplitude= sin(time)')
plt.grid(True, which ="both")
plt.show()
```

```python
time =np.arange(0,20,0.2)
amplitude =np.sin(time)
plt.plot(time, amplitude)
plt.title('Sine wave')
plt.xlabel('Time')
plt.ylabel('Amplitude= sin(time)')
plt.grid(True, which ="both")
plt.show()
```

Cos waves

```python
time = np.arange(0,20,0.1)
amplitude = np.cos(time)
plt.plot(time, amplitude)
plt.title('cos wave')
plt.xlabel('Time')
plt.ylabel('Amplitude= cos(time)')
plt.grid(True, which ="both")
plt.show()
```

Tan wave

```python
time =np.arange(0,20,0.2)
amplitude=np.tan(time)
plt.plot(time, amplitude)
plt.title('Tangent wave')
plt.xlabel('Time')
plt.ylabel('Amplitude= tan(time)')
plt.grid(True, which ="both")
plt.show()
```
